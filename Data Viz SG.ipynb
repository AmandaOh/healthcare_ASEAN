```python
%matplotlib inline
import numpy as np
import os
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# Load a stored pickle file as a pandas DataFrame
def load_pickle_as_df(pklname):
    pickle_folder = "./pickle_output"
    if os.path.exists(os.path.join(pickle_folder, pklname)):
        return pd.read_pickle(os.path.join(pickle_folder, pklname))
```

```python
malaria_data = load_pickle_as_df("malaria_data.pkl")
dengue_fever_data = load_pickle_as_df("dengue_fever_data.pkl")
dengue_h_fever_data = load_pickle_as_df("dengue_h_fever_data.pkl")
```

## Plot trends of the disease data - number of cases on weekly basis

```python
colors = ['r', 'g', 'b', 'm','c']
markers = ['*', 's', 'p', '8','h']
```

## Malaria

```python
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5);
for i, year in enumerate(range(2012, 2017)):
    malaria_year_data = malaria_data[malaria_data['year']==year]
    plt.plot(malaria_year_data.week, malaria_year_data.number_of_cases, color = colors[i], marker=markers[i], label=str(year))
    plt.legend()
    plt.xlabel('Week number')
    plt.ylabel('Number of cases')
    plt.title('Malaria weekly trend')
```

## Dengue Fever

```python
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5);
for i, year in enumerate(range(2012, 2017)):
    denguef_year_data = dengue_fever_data[dengue_fever_data['year']==year]
    plt.plot(denguef_year_data.week, denguef_year_data.number_of_cases, color = colors[i], marker = markers[i], label=str(year))
    plt.legend()
    plt.xlabel('Week number')
    plt.ylabel('Number of cases')
    plt.title('Dengue Fever weekly trend')
```

## Dengue Haemorrhagic fever

```python
fig = plt.gcf()
fig.set_size_inches(18.5, 10.5);
for i, year in enumerate(range(2012, 2017)):
    dengueh_year_data = dengue_h_fever_data[dengue_h_fever_data['year']==year]
    plt.plot(dengueh_year_data.week, dengueh_year_data.number_of_cases, color = colors[i], marker = markers[i], label=str(year))
    plt.legend()
    plt.xlabel('Week number')
    plt.ylabel('Number of cases')
    plt.title('Dengue Haemorrhagic Fever weekly trend')
```
